```python
import pandas as pd
import sklearn
from sklearn import linear_model
path = "C:\\Users\\spoorthi s jain\\Desktop\\Internship\\machine learning\\2022projects\\Logistic\\Mobilepricerange prediction classification\\train.csv"
data = pd.read_csv(path)
print(data)
print(data.info())
output=data.drop(['battery_power','blue','clock_speed','dual_sim','fc','four_g','int_memory','m_dep','mobile_wt','n_cores','px_height','px_width','ram','sc_h','sc_w','talk_time','three_g','touch_screen','wifi','pc'],'columns')  
inputs=data.drop(['price_range', 'pc'],'columns')
print(inputs)
print(output)
model=linear_model.LogisticRegression()
model.fit(inputs,output)
```

```python
print(model.score(inputs, output))
```
